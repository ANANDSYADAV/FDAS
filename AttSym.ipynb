```python
import os
import cv2
from deepface import DeepFace

# Path to the dataset
dataset_path = 'C:/Users/anand/Desktop/Attendance System/dataset'
```

```python
# Loop through the dataset folders
for student_folder in os.listdir(dataset_path):
    student_path = os.path.join(dataset_path, student_folder)
    
    if os.path.isdir(student_path):
        for image_name in os.listdir(student_path):
            image_path = os.path.join(student_path, image_name)
            
            # Load image using OpenCV
            img = cv2.imread(image_path)
            
            # Run DeepFace detection and analysis
            try:
                # Perform face detection using specified backend (e.g., 'mtcnn')
                result = DeepFace.analyze(img, actions=['emotion', 'gender'], detector_backend='mtcnn')
                print(f"Analyzing {image_name}: {result}")
            except Exception as e:
                print(f"Error processing {image_name}: {e}")
```

```python
import cv2
from deepface import DeepFace

# Load the webcam (0 is the default camera, change to 1 if you have multiple cameras)
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open video stream or file")
    exit()

while True:
    # Capture frame-by-frame from the camera
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    # Convert the frame (if necessary) or pass the frame directly to DeepFace for recognition
    try:
        # Perform face recognition with DeepFace
        result = DeepFace.find(frame, db_path=r'C:\Users\anand\Desktop\Attendance System\dataset', detector_backend='opencv')

        if len(result) > 0:
            # Assuming result[0] contains the first detected face in the database
            recognized_face = result[0]["identity"][0]
            # Draw a rectangle around the face (optional, if detection info available)
            # cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

            # Display the recognized name on the frame
            cv2.putText(frame, recognized_face, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    except Exception as e:
        # If DeepFace fails to detect or recognize a face, print the error
        print(f"Error: {e}")

    # Display the resulting frame
    cv2.imshow('Attendance System - Face Recognition', frame)

    # Break the loop if the user presses the 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
